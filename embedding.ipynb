```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[ 'the glass of milk',
      'the glass of juice',
      'the cup of tea',
      'I am a good boy',
      'I am a good developer',
      'understand the meaning of words',
      'your videos are good',]
```

```python
sent
```

```python
#define the vocabulary size
voc_size=10000
```

```python
sent
```

```python
#one hot representation
one_hot_repr=[one_hot(words,voc_size)for words in sent]
one_hot_repr
```

```python
#word embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential

```

```python
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length) #zeroes in beginning and you can also write post #to convert all the sentences into same length means all the sentences will be of equal words
print(embedded_docs)
```

```python
#feature representation
dim=10 #dimension is 10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
